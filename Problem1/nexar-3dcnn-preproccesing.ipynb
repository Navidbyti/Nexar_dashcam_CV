```python
import os
import numpy as np
import gc
from google.colab import drive

# ==========================================
# CONFIG
# ==========================================
old_folder = "/content/drive/MyDrive/Nexar_event_batches_LSTM"
new_folder = "/content/drive/MyDrive/Nexar_event_batches_3DCNN"

# ==========================================
# MOUNT DRIVE
# ==========================================
drive.mount('/content/drive')

if not os.path.exists(new_folder):
    os.makedirs(new_folder)

# ==========================================
# GET ALL OLD BATCH FILES
# ==========================================
batch_files = sorted([
    f for f in os.listdir(old_folder)
    if f.endswith(".npz")
])

print("Found batches:", len(batch_files))

# ==========================================
# PROCESS EACH BATCH
# ==========================================
for batch_file in batch_files:

    print(f"\nProcessing {batch_file}")

    old_path = os.path.join(old_folder, batch_file)
    data = np.load(old_path)

    X = data["X"]  # (N, 9, 224, 224, 3)
    y = data["y"]

    # --------------------------------------
    # Compute frame differences
    # --------------------------------------
    # diff[t] = frame[t+1] - frame[t]
    X_diff = X[:, 1:] - X[:, :-1]  # (N, 8, 224, 224, 3)

    print("Old shape:", X.shape)
    print("New shape:", X_diff.shape)

    # --------------------------------------
    # Save new batch
    # --------------------------------------
    new_path = os.path.join(new_folder, batch_file)

    np.savez(
        new_path,
        X=X_diff.astype(np.float32),
        y=y
    )

    # --------------------------------------
    # Free memory
    # --------------------------------------
    del X
    del X_diff
    del data
    gc.collect()

print("\n✅ All batches converted for 3D CNN.")
```
